```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')
df = pd.read_csv('/content/gdrive/MyDrive/TseverlesenData1.csv')
```

```python
df['Цалин'] = df['Цалин'].str.replace(',','')
df['Цалин'] = df['Цалин'].str.replace('-ааc дээш','')
df['Цалин'] = df['Цалин'].str.replace('-',' ')
num = df['Цалин'].str.split(" ")
for i in range(len(num)):
  if num[i][0]!='':
    df['Цалин'][i] = int((int(num[i][0].strip())+int(num[i][0].strip()))/2)
df
year = []
month = []
day = []
count = []
dates = df['Зар нийтэлсэн огноо'].str.split("/")
for i in range(len(dates)):
  month.append(dates[i][0])
  day.append(dates[i][1])
  count.append((int(dates[i][0])-1)*30+int(dates[i][1]))
  year.append(dates[i][2].split(" ")[0])
df.insert(10, "Жил", year, True)
df.insert(11, "Сар", month, True)
df.insert(12, "Өдөр", day, True)
df.insert(13, "Хоног", count, True)
```

```python
#
df['Цалин'] = df['Цалин'].astype(int)
sorted = df.sort_values(by =["Цалин"])
g = sns.relplot(x='Салбар', y = 'Цалин', data = sorted, height = 8)
g.set_xticklabels(rotation=90)
plt.show()
```

```python
#Ажлын заруудаас ямар хэмжээний цалин санал болголт их байгааг харуулах график
fig, ax = plt.subplots(figsize=(50, 20))
df['Цалин'] = df['Цалин'].astype(int)
sorted = df.sort_values(by =["Цалин"])
sns.histplot(sorted, kde=True, bins = 15)
plt.show()
```

```python
#Бүтэн цагийн, ээлжийн, улирлын, цагийн ажил салбаруудад хэр хэмжээтэй байгааг харуулах
fig, ax = plt.subplots(figsize=(40, 20))
sns.countplot(x="Салбар", data=df, hue = "Төрөл")
plt.xticks(rotation=90)
plt.show()
```

```python
#Хамгийн их ажлын зар оруулсан 5 комданиудын цалингийн тархалт
most_companies = df['Компани'].value_counts().head(5).index.tolist()
new_df = df[df['Компани'].isin(most_companies)]
new_df['Цалин'] = new_df['Цалин'].astype(int)
sorted = new_df.sort_values(by =["Цалин"])
g = sns.relplot(data=sorted, x="Салбар", y="Цалин", hue="Компани", style="Компани", height = 7)
g.set_xticklabels(rotation=-90)
```

```python
sns.jointplot(data=df, x="Цалин", y="Сар")
```

```python
#Ажлын зар орсон хугацаа болон цалингийн тархалтын хамаарал
fig, ax = plt.subplots(figsize=(7, 5))
sorted = df.sort_values(by =["Цалин"])
sns.boxplot(x="Цалин", y="Сар", data=sorted, palette="YlGnBu")
plt.show()
sns.catplot(
    data=df,
    x="Сар", y="Цалин", kind="boxen",
)
```

```python
#Цалин тархалт
from matplotlib import colors
df['Цалин'] = df['Цалин'].astype(int)
fig, axs = plt.subplots(3, 1, figsize=(5, 15), sharex=True, sharey=True, tight_layout=True)
axs[0].hist2d(df['Цалин'], df['Цалин'], bins=40)
axs[1].hist2d(df['Цалин'], df['Цалин'], bins=40, norm=colors.LogNorm())
axs[2].hist2d(df['Цалин'], df['Цалин'], bins=(80, 10), norm=colors.LogNorm())
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
df['Хоног'] = df['Хоног'].astype(int)
sorted = df.sort_values(by =["Хоног"])
sns.histplot(sorted, kde=True, bins = 65)
plt.show()
```

```python
#хамгийн эрэлттэй байгаа 5н ажлын байрны цалин
most_job = df['Салбар'].value_counts().head(5).index.tolist()
new_df = df[df['Салбар'].isin(most_job)]
new_df['Цалин'] = new_df['Цалин'].astype(int)
sorted = new_df.sort_values(by =["Цалин"])
sns.set_style("whitegrid")
g = sns.relplot(data=sorted, x="Хоног", y="Цалин", hue="Салбар", kind = "line", height = 7)
sns.despine(top=True, right=True)
g.ax.patch.set_facecolor('white')
g.set_xticklabels(rotation=-90)
```

```python
#Ажлын салбарууд түвшинөөс хамаарч ямар цалин авч буйг харуулах.
df['Цалин'] = df['Цалин'].astype(int)
fig, ax = plt.subplots(figsize=(10, 10))
sns.set_style("darkgrid")
data_agg = df.groupby(['Салбар', 'Түвшин']).mean().reset_index()
pivoted_data = data_agg.pivot('Салбар', 'Түвшин', 'Цалин')
sns.heatmap(pivoted_data, cmap='coolwarm')
plt.show()
```
